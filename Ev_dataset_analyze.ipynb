```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
if 'Month' in data.columns:
    data['Month'] = pd.to_datetime(data['Month'], format='%b-%y')
    data.set_index('Month', inplace=True)

plt.figure(figsize=(14, 7))
data.plot.area(alpha=0.5)
plt.title('Cumulative Vehicle Registrations by Type')
plt.ylabel('Cumulative Count of Vehicles')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 10))
g = sns.FacetGrid(data.reset_index(), col="variable", col_wrap=3, height=4, aspect=1)
g.map(sns.lineplot, "Month", "value", marker='o')
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Month", "Count of Vehicles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

data.reset_index(inplace=True)
plt.figure(figsize=(14, 7))
data.set_index('Month').plot(kind='bar', stacked=True)
plt.title('Stacked Vehicle Registrations by Month')
plt.ylabel('Count of Vehicles')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.legend(title='Vehicle Type')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 6))
plt.scatter(data['DIESEL'], data['Petrol'], alpha=0.6)
plt.title('Scatter Plot: Diesel vs Petrol Registrations')
plt.xlabel('Diesel Registrations')
plt.ylabel('Petrol Registrations')
plt.grid()
plt.tight_layout()
plt.show()
plt.figure(figsize=(14, 7))
sns.kdeplot(data=data, fill=True)
plt.title('Density Plot of Vehicle Registrations')
plt.xlabel('Count of Vehicles')
plt.ylabel('Density')
plt.legend(title='Vehicle Type')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
data = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
kmeans = KMeans(n_clusters=4)
data['Cluster'] = kmeans.fit_predict(X)
plt.scatter(data['Feature1'], data['Feature2'], c=data['Cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
